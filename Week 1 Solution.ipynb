<h3> Q1

```python
## (1) 행렬 A의 최소제곱해 구하기 
## 편의상 파이썬 이용해서 구하겠습니다.

import numpy as np

## Ax = b의 형태 만들기
A = np.array([[1, 3, 5], [1, 1, 0], [1, 1, 2], [1, 3, 3]])
b = np.array([2, 3, -1, 2])

## 정규방정식은 AT * Ax = AT * b이므로 양변에 a의 전치행렬 곱하기
AtA = np.dot(A.T, A)
Atb = np.dot(A.T, b)

print('좌변(AtA)의 형태\n', AtA)
print('우변(Atb)의 형태\n', Atb)

## 즉, 4x1 + 8x2 + 10x3 = 6, 8x1 + 20x2 + 26x3 = 14, 10x1 + 26x2 + 38x3 = 14를 풀어야함.
## 방정식 해 구하기 -> numpy에서는 방정식의 해를 구할 수 있도록 하는 메서드 존재

x = np.linalg.solve(AtA, Atb)
print('최소제곱해 : ', x) ## (0, 2, -1)


##======================================================##
# (2) |b-Ax| 구하기 
Ax = np.dot(A, x)
print('최소제곱해에서 오차로 이루어진 행렬 : ', b - Ax)

error = np.linalg.norm(b-Ax)
print('최소제곱해에서의 오차 : ', error) ## 최종 오차 2
```

<h3> Q2 </h3>

```python
import numpy as np

## 모든 값이 0인 100 x 2 행렬 X 만들기 
X = np.zeros((100, 2))

## 첫 번째 열과 두 번째 열에 각각 요구한 대로 값 집어넣기
X[:, 0] = np.random.randint(10, 100, 100)
X[:, 1] = np.random.randint(100, 2000, 100)

print('첫 번째 열과 두 번째 열에 들어간 결과\n')
print(X[:, 0])
print()
print(X[:, 1])
print()

## X와 동일한 크기를 갖는 행렬 만들기 (standard_x)
standard_x = np.zeros_like(X)
print('standard_x : ', standard_x.shape)

## X를 표준정규분포로 표준화 하기
standard_x[:, 0] = (X[:, 0] - X[:, 0].mean()) / X[:, 0].std()
standard_x[:, 1] = (X[:, 1] - X[:, 1].mean()) / X[:, 1].std()

column_0_mean, column_0_std = standard_x[:, 0].mean(), standard_x[:, 0].std()
column_1_mean, column_1_std = standard_x[:, 1].mean(), standard_x[:, 1].std()

print('표준화된 X : ', standard_x)
print('표준화 결과 확인하기 -> {}, {}, {}, {}'.format(column_0_mean, column_1_mean, column_0_std, column_1_std))
```

```python
data_arr = np.loadtxt('population.csv', delimiter = ',', dtype = np.int64)
print("[Data Loaded]:", data_arr.shape, '\n')
```

<h4> (1) 연령대별 남녀 성비 출력 </h4>

```python
## 기준이 20대, 30대, 40대, 50대이므로 lookup table 생성
### 코드 동작 방식 -> 첫 번째는 20대 미만이니까 0 ~ 20, 마지막은 50대 이상이니까 50 ~ 150
lookup_table = np.array([0, 20, 30, 40, 50, 150])
title = ["20대미만\t", "20대\t", "30대\t", "40대\t", "50대이상\t"]

print('\t\t남성\t여성')
for i, lookup_age in enumerate(title):

    # row_ages : 조건에 맞는 연령대를 슬라이싱 기법 및 불린 인덱싱 이용하여 추출
    # boys_ratio : 연령대별로 추출 후에 남성 비율 측정
    # 남성 비율 측정 후 여성 비율은 1 - boys_ratio의 형태로 계산
    row_ages = data_arr[(data_arr[:, 1] >= lookup_table[i]) & (data_arr[:, 1] < lookup_table[i+1])]
    boys_ratio = len(row_ages[row_ages[:, 0] == 1]) / len(row_ages)
    print(title[i] +'\t{:.2f}\t{:.2f}'.format(boys_ratio, 1 - boys_ratio))
```
